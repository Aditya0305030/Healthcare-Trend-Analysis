```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.common.exceptions import NoSuchElementException
import time
import pandas as pd
```

```python
# Run in headless mod
driver=webdriver.Chrome(service=Service(ChromeDriverManager().install()))
```

```python
search_query = "AI in healthcare"
url = f"https://pubmed.ncbi.nlm.nih.gov/?term={search_query.replace(' ', '+')}"
driver.get(url)
time.sleep(3)
```

```python
titles = []
dates = []

articles = driver.find_elements(By.CLASS_NAME, 'docsum-content')
```

```python
for article in articles:
    try:
        title = article.find_element(By.CLASS_NAME, 'docsum-title').text
        date = article.find_element(By.CLASS_NAME, 'docsum-journal-citation').text
        titles.append(title)
        dates.append(date)
    except Exception as e:
        print("Error:", e)
```

```python
search_categories = {
    "Medical Research": ["clinical trial", "medical research", "peer-reviewed"],
    "Treatments": ["drug treatment", "therapy", "vaccine"],
    "Disease Outbreaks": ["pandemic", "epidemic", "infectious disease"],
    "Technology Advancements": ["AI in healthcare", "telemedicine", "robotic surgery"],
    "Healthcare Investments": ["healthcare startup", "health investment", "biotech funding"]
}

all_data = []
max_pages = 1

for category, keywords in search_categories.items():
    for keyword in keywords:
        url = f"https://pubmed.ncbi.nlm.nih.gov/?term={keyword.replace(' ', '+')}"
        driver.get(url)
        time.sleep(2)
        current_page = 1

        while current_page <= max_pages:
            print(f"Scraping page {current_page}")
            time.sleep(2)
        
        articles = driver.find_elements(By.CLASS_NAME, 'docsum-content')
        for article in articles:
            try:
                title = article.find_element(By.CLASS_NAME, 'docsum-title').text
                citation = article.find_element(By.CLASS_NAME, 'docsum-journal-citation').text
                all_data.append({
                    "Category": category,
                    "Keyword": keyword,
                    "Title": title,
                    "Citation": citation
                })
            except Exception as e:
                print(f"Error: {e}")
        try:
                # Click the "Next" button
                next_button = driver.find_element(By.XPATH, '//button[@class="next-page-btn"]')
                driver.execute_script("arguments[0].click();", next_button)
                current_page += 1
                time.sleep(2)
        except NoSuchElementException:
                print(" No more pages.")
                break
```

```python
search_categories = {
    "Medical Research": ["clinical trial", "medical research"],
    "Treatments": ["drug treatment", "therapy"],
    "Disease Outbreaks": ["pandemic", "infectious disease"],
    "Technology Advancements": ["AI in healthcare", "telemedicine"],
    "Healthcare Investments": ["healthcare startup", "health investment"]
}

max_pages = 3  # number of pages to scrape per keyword
all_data = []

for category, keywords in search_categories.items():
    for keyword in keywords:
        print(f"\n Scraping '{keyword}' under category '{category}'")
        base_url = f"https://pubmed.ncbi.nlm.nih.gov/?term={keyword.replace(' ', '+')}"
        driver.get(base_url)
        time.sleep(2)

        current_page = 1

        while current_page <= max_pages:
            print(f"Scraping page {current_page}...")
            time.sleep(2)

            articles = driver.find_elements(By.CLASS_NAME, 'docsum-content')
            for article in articles:
                try:
                    title = article.find_element(By.CLASS_NAME, 'docsum-title').text
                    citation = article.find_element(By.CLASS_NAME, 'docsum-journal-citation').text
                    all_data.append({
                        "Category": category,
                        "Keyword": keyword,
                        "Title": title,
                        "Citation": citation
                    })
                except Exception as e:
                    print(f"Error parsing article: {e}")

            try:
                next_button = driver.find_element(By.XPATH, '//button[@class="next-page-button"]')
                driver.execute_script("arguments[0].click();", next_button)
                current_page += 1
            except NoSuchElementException:
                print(" No more pages available.")
                break
```
